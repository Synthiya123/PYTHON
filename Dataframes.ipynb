```python
Employees=['Cruz','Clarie','cinthiya','Darlene','Giftson','Judson']
salary=[35000,50000,25000,80000,75567,65432]

emp_details = pd.DataFrame(
    {
        "Names": Employees,
        "salary P/M": salary
     }
     
  )
emp_details
```

#Dealing with missing values

```python

#Dataframes have specific commands which help with the handling of missing data

sparse = pd.DataFrame({'a': [1, 2, 3],
                       'b': [np.nan, 4, np.nan],
                       'c': [5, np.nan, np.nan]})
sparse
```

```python
sparse.isna()
```

```python
sparse.fillna(0) 
```

```python
sparse.dropna(axis=1)

```

```python
contacts = pd.DataFrame(
    {
        'Names' : ['shiny', 'sharma'],
        'tel'  : ['+44 1234 5678', '+44 3245 5564']
    }
)
contacts
```

```python
emp_details.merge(contacts, left_on='Names', right_on='Names')   # defaults to inner join
```

```python
emp_details.merge(contacts, left_on='Names', right_on='Names', how='outer')
```

```python
emp_details.merge(contacts, left_on='Names', right_on='Names', how='left')
```

```python
emp_details.merge(contacts, left_on='Names', right_on='Names', how='right')
```

```python
emp_details.merge(contacts,how='cross')
```
