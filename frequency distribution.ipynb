VISUALISATION - FREQUENCY DISTRIBUTION.

```python
sns.histplot(data=unemp,x='2020')
```

```python
twy = pd.crosstab(index=unemp['2020'],columns='f')
twy
```

```python
sns.histplot(data=unemp,x='2021',color='r').set(title="YEAR 2021")
plt.show()
```

```python
tyo = pd.crosstab(index=unemp['2021'],columns='freq')
tyo
```

```python
sns.histplot(data=unemp,x='1995',color='y')
```

```python
nifi = pd.crosstab(index=unemp['1995'],columns='freq')
nifi
```

```python
fig,axes= plt.subplots(nrows=1,ncols=3,figsize=(30,8))
sns.histplot(x='1994',data=unemp,ax=axes[0],color='seagreen')
axes[0].set_title('Year 1994')
axes[0].set(xlabel='x1')

sns.histplot(x='1996',data=unemp,ax=axes[1],color='crimson')
axes[1].set_title('Year 1996')
axes[1].set(xlabel='x2')

sns.histplot(x='2000',data=unemp,ax=axes[2],color='peru')
axes[2].set_title('Year 2000')
axes[2].set(xlabel='x3')
plt.show()
```

```python
mulp=pd.crosstab(index=unemp['1994'],columns='freq94')
mulp
```

```python
nisi=pd.crosstab(index=unemp['1996'],columns='freq96')
nisi
```

```python
twth=pd.crosstab(index=unemp['2000'],columns='freq20')
twth
```

```python
fig,axes= plt.subplots(nrows=1,ncols=3,figsize=(30,8))
sns.distplot(unemp['1993'],ax=axes[0],color='tomato')
axes[0].set_title('Year 1993')
axes[0].set(xlabel='x1')

sns.distplot(unemp['1996'],ax=axes[1],color='orchid')
axes[1].set_title('Year 1996')
axes[1].set(xlabel='x2')

sns.distplot(unemp['1999'],ax=axes[2],color='steelblue')
axes[2].set_title('Year 1999')
axes[2].set(xlabel='x3')
plt.show()
```
