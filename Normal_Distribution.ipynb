**DISTRIBUTION PLOTS**

```python
sns.distplot(heart['trtbps'])
```

```python
sns.kdeplot(heart['trtbps'])
```

Blood Pressure level is observed to be normally distributed and right skewed.

```python
heart['trtbps'].skew(axis = 0, skipna = True)
```

```python
heart.skew(axis=0,skipna=True)
```

```python
pskew=(heart['trtbps'].mean()-heart['trtbps'].median())/heart['trtbps'].std()
pskew
```

```python
sns.histplot(data=heart,x='chol',color='red')
```

```python
sns.kdeplot(data=heart,x='chol')
```

```python
heart['chol'].skew(axis = 0, skipna = True)
```

The variable chlosterol is also normal distributed.

```python
sns.jointplot(x='thalachh',y='age',data=heart,kind='kde',color='y')
```

Comparing two variables: thalachh(max heart rated achieved) and age


while comparing the age with thalachh, thalachh is normally distributed with left skewness whereas age is not distributed normally.
