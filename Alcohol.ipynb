**9.Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)**

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False
```

```python
stu_alco['legal_drinker'] = stu_alco['age'].apply(majority)
stu_alco.head()
```
