TO delete the unnamed column

```python
tot_bil=sns.distplot(tips.total_bill)
```

```python
tot_bil.set(xlabel = 'Value', ylabel = 'Frequency', title = "Total Bill")
```

```python
sns.despine()
```

To create the relationship of 2 or more colums in a single plot 

```python
sns.pairplot(tips)
```

To represent the relationship between 2 columns

```python
sns.jointplot(x ="total_bill", y ="tip", data = tips,color='purple')
```

#Relationship between days and total bill

```python
sns.stripplot(x = "day", y = "total_bill", data = tips)
```

# To create the scatter plot and differentiate it by 'hue'

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True);
```



*   Differentiate using Boxplot



```python
sns.boxplot(y= "total_bill", x = "day", hue = "time", data = tips)
```

# To create 2 histograms side by side

```python
sns.set(style = "ticks")
```

```python
from seaborn.utils import plt
h= sns.FacetGrid(tips, col = "time")
h.map(plt.hist, "tip")
```

```python
scat = sns.FacetGrid(tips, col = "sex", hue = "smoker")
scat.map(plt.scatter, "total_bill", "tip", alpha =.9)

scat.add_legend();
```
